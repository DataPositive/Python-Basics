#THEORY QUESTIONS

1. What is Python, and why is it popular?
- Python is a simple, beginner-friendly programming language used for tasks like web development, data analysis, machine learning, and automation. It is popular because it is easy to learn, works for many purposes, has a lot of ready-made libraries, a big supportive community, and runs on all major platforms like Windows, macOS, and Linux.

2. What is an interpreter in Python?
- An interpreter in Python is a program that reads and executes Python code line by line. It converts the code into machine-readable instructions and runs it immediately, making Python an interpreted language.

3. What are pre-defined keywords in Python
- Pre-defined keywords in Python are reserved words that have special meanings and cannot be used as variable names. Examples: `if`, `else`, `for`, `while`, `True`, `False`, `def`, and `return`.

4. Can keywords be used as variable names
- No

5. What is mutability in Python
- Mutability in Python refers to whether an object can be changed after creation. Mutable objects (eg. lists) can be altered, while immutable objects (eg, strings) cannot.

6. Why are lists mutable, but tuples are immutable?
-  Lists are made to be flexible, so you can change their items. Tuples are meant to be fixed, which makes them faster and safer when you don’t want the data to change

7. What is the difference between “==” and “is” operators in Python?
- “==” checks if two values are the same, while “is” checks if two variables refer to the same object in memory.

8. What are logical operators in Python?
- Logical operators like and, or, and not are used to combine multiple conditions.

9. What is type casting in Python
- Type casting means converting one data type to another, like turning an integer into a string or a float.

10. What is the difference between implicit and explicit type casting?
- Implicit casting happens automatically by Python, while explicit casting needs functions like int(), float(), or str() to change types

11. What is the purpose of conditional statements in Python?
- Conditional statements like if, elif, and else let Python decide which part of the code to run based on whether a condition is True or False

12. How does the elif statement work?
 - The elif statement checks another condition if the previous if or elif was False.

13. What is the difference between for and while loops?
- A for loop is used when you know how many times you want to repeat something, while a while loop runs as long as a condition is true.

14. Describe a scenario where a while loop is more suitable than a for loop
- A while loop is useful when you don’t know how many times you need to loop, like waiting for the user to enter valid input or running a process until it finishes.

Q9. Write a Python program to convert user input from string to integer, float, and boolean types

```python
user_input = input("Enter a value: ")

# Convert to integer
try:
    int_value = int(user_input)
    print(f"Integer: {int_value}")
except ValueError:
    print("Cannot convert to integer.")

# Convert to float
try:
    float_value = float(user_input)
    print(f"Float: {float_value}")
except ValueError:
    print("Cannot convert to float.")

# Convert to boolean
try:
    bool_value = bool(user_input)
    print(f"bool: {bool_value}")
except ValueError:
    print("Cannot check the bool.")
```

Q14. Write a program to reverse a string using a while loop

```python
strng = input("enter a string: ")
i = len(strng) - 1
while i >= 0:
    print(strng[i], end="")
    i -= 1
```

Q15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
a = int(input("enter a number: "))
f = 1
while a > 0:
    f *= a
    a -= 1
print(f)
```
